## Introduction
In this assignment, we will have a look at distributed denial of service (DDoS) attacks, which are attacks that can take services and infrastructures out of service by overwhelming them (Santanna et al., 2018). As Douglas et al. (2017) explain, it is relatively easy to carry out a DDoS-attack without having knowledge of the technical aspects of it. The reason for this is that everyone can easily hire a service that performs the attack for you because of the existence of booter websites. ‘The provision of booter services is maintained by the “easy money,” with little cost in terms of time spent maintaining the sites’ (Hutchings & Clayton, 2016, p. 1176). These websites maintain botnets to use them for DDoS attacks. Some of these botnets are usually shared among different booters in case the owner of the booters is the same. Users just have to access a web front end and pay for a plan on the website and specify a target and time of attack and it is that easy (Kerner, S. M., August 5, 2013).  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;One of the people that have dealt with Booter websites in the context of DDOS attacks is Brian Krebs, a freelance journalist who wrote for The Washington Post and the Security Fix blog. On his website (krebsonsecurity.com), he writes about topics related to security and cybercrime, such as online fraud, online stalking, spamming, hacking, and so on. He investigated a booter website called booter.tw and published the internal database of the booter website on his blog website (Krebs, March 13, 2018). This database will be used in this assignment for research purposes. A closer look will be taken at number and intensity of the attacks, the victims and the timing of the attacks.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;The goal of this assignment is to gain an overview of the extent to which DDOS attacks using booters occur. By gathering information about who attacks, how long attacks take and who the victims are, awareness can be created in the population. Moreover, the analysis may help provide first steps for possible solutions to solve the problem.

## Questions

### What is the total number of attack records?
To anwser this question we executed the following SQL query:
```sql
SELECT COUNT(*) as num_attacks 
    FROM attacks;
``` 
The result of this query was: <b>48.844</b>.  

The first attack in the database was executed on January 24 2013, the last attack was executed on March 15 2013. This is a timespan of only <b>50</b> days. This means that, on average, every day around <b>1000</b> attacks were executed.

To anwser this question we executed the following SQL query:

The result of this query was: <b>277</b>.  
When we look at the total amount of attacks, this would mean that evey user (on average) executed around <b>176</b> attacks.

### What are the top 10 victim IP addresses?
To anwser this question we executed the following SQL query:
```sql
SELECT COUNT(*) AS num_attacks, ip  
    FROM attacks  
    WHERE ip REGEXP '([0-9]{1,3}\.){3}[0-9]{1,3}'  
    GROUP BY ip
    ORDER BY num_attacks DESC
    LIMIT 10
``` 
This query counts the number of attacks per ip address. We use the `REGEXP` clause to filter out all entries which do not have a valid ip address.

````python
attacks_per_ip = pd.read_csv('data/attacks_per_ip.csv')
bar_chart(attacks_per_ip['num_attacks'], attacks_per_ip['ip'], 'Number of attacks for top 10 victim ip addresses')
````

### What are the top 10 victim domain names?
To anwser this question we executed the following SQL query:
```sql
SELECT COUNT(*) AS num_attacks, ip  
  FROM attacks  
  WHERE ip NOT REGEXP '([0-9]{1,3}\.){3}[0-9]{1,3}'  
  GROUP BY ip
  ORDER BY num_attacks DESC
  LIMIT 10
``` 
This query is almost the same as the one for the ip adresses. The only difference being that the `REGEXP` clause is now negated. We will now only get records which are not a valid ip address, but most likely are a valid domain name.

````python
attacks_per_domain = pd.read_csv('data/attacks_per_domain.csv')

bar_chart(attacks_per_domain['num_attacks'], attacks_per_domain['domain'], 'Number of attacks for top 10 victim domain names')
````

As we can see the top 3 used attack types are: `ESSYN`, `UDP`, and `RUDY`. These 3 together make up almost 75% of the total attacks. 

#### How do these different types of attack work?

##### UDP
A UDP server attack is one of the most basic kinds of DDoS attack. It works as follows: A botnet of internet-connected devices will be sending UDP packets to the victim’s server. If a large enough botnet executes this procedure the victim’s server will be overwhelmed by the amount of packets to route through(UDP flood attack). However, there are ample options to limit the scale of the botnet to be used. An example is by utilizing dns amplification to greatly reduce the device needed.

##### ESSYN
ESSYN attack type is the one which was used the most. That is because it consumes the resources of the target much faster than the other attack types. With a SYN flood attack, the attackers send TCP connection requests faster than the targeted machine can handle, which results in Denial of Service (Imperva).

##### RUDY
Rudy is short for “R U Dead Yet” and is categorized as a “low and slow” attack. It works by initiating a HTTP GET request and not completing it. This is done by sustaining the connection with the server by sending arbitrarily small packets, often even 1Kb. When opening several of these connections the server slowly runs out of memory (Damon, Dale, Laron., Mache, Land, & Weiss).

### On which hours of the day were the most attacks executed?
To anwser this question we executed the following SQL query:
```sql
SELECT COUNT(*) AS num_attacks, HOUR(FROM_UNIXTIME(time)) AS hour 
  FROM attacks 
  GROUP BY hour 
  ORDER BY hour
```

In the graph below we can clearly see that the amount of attacks is the lowest in the morning. During the day, the amount of attacks slowly increases until the highest point around 12pm.

````python
atph = pd.read_csv('data/attacks_per_hour.csv')
bar_chart(atph['num_attacks'], atph['hour'], 'Number of attacks per hour of the day')
````

### Does the number of attacks in each hour differ per day?

In the graph above, we can clearly see a trend in the number of attacks on each hour of the day. This graph aggregates the data of all days into one graph. However, this trend might vary per day. The number of attacks is higher on firday and saturday, so maybe also the trend on which hours most attacks are executed differ per day.

In the plots below we look at the attacks in each hour of the dat, for each day in the week.


````python
fig = plt.figure(constrained_layout=True, figsize=(12, 10))
grid = gridspec.GridSpec(ncols=2, nrows=4, figure=fig)

fig.add_subplot(grid[0, 0])
attacks_per_hour = pd.read_csv('data/attacks_per_hour_monday.csv')
bar_chart(attacks_per_hour['num_attacks'], attacks_per_hour['hour'], 'Number of attacks per hour on Monday', False)

fig.add_subplot(grid[0, 1])
attacks_per_hour = pd.read_csv('data/attacks_per_hour_tuesday.csv')
bar_chart(attacks_per_hour['num_attacks'], attacks_per_hour['hour'], 'Number of attacks per hour on Tuesday', False)

fig.add_subplot(grid[1, 0])
attacks_per_hour = pd.read_csv('data/attacks_per_hour_wednesday.csv')
bar_chart(attacks_per_hour['num_attacks'], attacks_per_hour['hour'], 'Number of attacks per hour on Wednesday', False)

fig.add_subplot(grid[1, 1])
attacks_per_hour = pd.read_csv('data/attacks_per_hour_thursday.csv')
bar_chart(attacks_per_hour['num_attacks'], attacks_per_hour['hour'], 'Number of attacks per hour on thursday', False)

fig.add_subplot(grid[2, 0])
attacks_per_hour = pd.read_csv('data/attacks_per_hour_friday.csv')
bar_chart(attacks_per_hour['num_attacks'], attacks_per_hour['hour'], 'Number of attacks per hour on Friday', False)

fig.add_subplot(grid[2, 1])
attacks_per_hour = pd.read_csv('data/attacks_per_hour_saturday.csv')
bar_chart(attacks_per_hour['num_attacks'], attacks_per_hour['hour'], 'Number of attacks per hour on Saturday', False)

fig.add_subplot(grid[3, 0])
attacks_per_hour = pd.read_csv('data/attacks_per_hour_sunday.csv')
bar_chart(attacks_per_hour['num_attacks'], attacks_per_hour['hour'], 'Number of attacks per hour on Sunday', False)

plt.show()
````

When looking at the graphs, we see that the trends for each day more or less follows the global trend. However, in the weekend, the graphs do change. On Saturday, the difference between the lowest and highest point is less than globally. On Sunday we see that the lowest point has shifted to a few hours later on the day. 

### What servers were used the most in the attacks?
To anwser this question we executed the following SQL query:
```sql
SELECT COUNT(*) AS num_attacks, s.name
    FROM attacks a
    INNER JOIN servers s ON s.ip = a.server 
    GROUP BY s.name
    ORDER BY num_attacks DESC
```

````python
most_used_servers = pd.read_csv('data/attacks_per_server.csv')
pie_chart(most_used_servers['num_attacks'], most_used_servers['server'])
````

For more then 90% of the attacks, only 3 servers were used.
