## Import Dependencies

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## Generate Bar Chart for # attacks per duration

```python
df = pd.read_csv('data/attack_duration.csv')
df.set_index('duration')[['num_attacks']].plot.bar()
plt.title('Number of attacks for each duration')
plt.show()
```

Only durations which have been executed >100 times. This is due to the dact that there are 1700 distinct durations. Showing them all would require a graph of a few pages big.

## Pie chart for most used attack types

```python
at = pd.read_csv('data/attack_types.csv')

fig1, ax1 = plt.subplots()
ax1.pie(at['num_attacks'], labels=at['type'], autopct='%1.1f%%',
        shadow=False, startangle=90)

#draw circle
centre_circle = plt.Circle((0,0),0.80,fc='white')
fig1.gca().add_artist(centre_circle)

# Equal aspect ratio ensures that pie is drawn as a circle
ax1.axis('equal')
plt.tight_layout()  
plt.show()
```

Grouped smallest 5 types into OTHER due to space left
