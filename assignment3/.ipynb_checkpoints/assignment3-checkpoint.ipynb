# Assignment 3: DDoS attack investigation

### Introduction
A Distributed Denial of Service (DDoS) attack can result in a blockade of a webserver and thus the denial of access to this website by its owner and clients (Behal & Kumar, 2017). The owner of the server and the clients of the server cannot use the server during the time of the attack, which may result in financial losses. It is important to gain more information about this phenomenon and how to complicate and detect it. As Behal and Kumar (2017, p. 96) ‘<i>it is crucial to detect such attacks in time to ensure the timely delivery of the widely used Internet-based services and applications</i>’. Carrying out a DDoS-attack is not that hard for an individual, because booter websites offer the service for money. Santanna et al. (2015) gave the example of a student attacking his own school by using such a booter website. The BBC reported in 2018 that the FBI approached DDoS attacks as a national threat and took several booter websites online (BBC, December 21 2018). These booter websites had been the source of many attacks and three people were arrested by the FBI.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;Because a DDoS-attack can lead to serious (financial) losses and carrying out these attacks is relatively easy, it is important to gain a better understanding of how such an attack is carried out, why, and by whom. The first step is to understand the attack itself. By analyzing the attack itself in a digital forensic way, this report tries to provide insight in the way attacks could possibly be investigated in the future. In order to do this, a DDoS attack was carried out on the network of the lecturer of the University of Twente. During the attack, the lecturer ran the programme Wireshark and published the data in the course group. The following report is based on this data.

### Results
The supplied trace exists of <b>117333</b> records. These records were recorded in a timespan of <b>71.872891</b> seconds.  

Although the supplied trace contains a trace of the attack, not all records in the trace are actually part of the attack. It also contains packets for i.e. the <i>ARP</i>, <i>DHCP</i>, and <i>DNS</i> protocols. Below we will briefly discuss, for these three protocols, what these packets do.

##### ARP
<i>ARP</i> is a request-response protocol used to find the <i>MAC</i> address which coresponds to an <i>IP</i> address. Every operating system has an <i>ARP</i> table. Every device on the network, with a operating system, has an <i>ARP</i> table. In this table, the device stores mappings of <i>IP</i> addresses to <i>MAC</i> addresses for other devices in the network. When a device wants to send something to an <i>IP</i> address, it needs to know the corresponding <i>MAC</i> address. If the device does not yet have the mapping for that <i>IP</i> address, it broadcasts an <i>ARP</i> request to find the <i>MAC</i> address for that <i>IP</i> address. The device to which the <i>IP</i> address belongs then responds with another <i>ARP</i> message which contains the <i>MAC</i> address of the device.  
In the screenshot below we can see ARP requests being sent. In the info section on the right side we see for which <i>IP</i> addresses the <i>MAC</i> address were requested, and by whom.
![ARP records + filter](screenshots/arp.png)

##### DNS
<i>DNS</i> is a protocol which is used to translate <i>Domain Names</i> into <i>IP</i> addresses. When a device wants to know the <i>IP</i> address which belongs to a <i>Domain Name</i>, it sends out an <i>DNS</i> request. This request is then sent to special <i>DNS</i> servers which will try to find the correct <i>IP</i> address for the <i>Domain Name</i>  
In the screenshot below we can see a slice of <i>DNS</i> packets from the trace. As we can see, these <i>DNS</i> requests are sent for domains of: <i>apple, spotify, icloud, and utwente</i>. The target PC of the attack was an Macbook, which explains the requests to <i>apple</i> and <i>icloud</i>. The <i>utwente</i> request exlpainable by the fact that the PC was located on the University of Twente. The requests to <i>spotify</i> can be explained by the fact that the spotify application was probably opened on the PC.
![DNS records + filter](screenshots/dns.png)

##### DHCP
<i>DHCP</i> is a protocol which is used to request an <i>IP</i> address. When a device connects to a new network, it does not yet have an <i>IP</i> address on that network. Therefore it sends out a <i>DHCP</i> request to the router asking for a new <i>IP</i> address. It is possible for the device to request a specific <i>IP</i> address. If the device does not request an specific <i>IP</i> address, it will be assigned a random <i>IP</i> address that is still free from the <i>DHCP</i> pool.
![DHCP records + filter](screenshots/dhcp.png)


To focus on the attack itself, the following filter was used in Wireshark: `tcp.dstport == 80 && ip.dst == 192.168.0.100`. This filter was build because the target <i>IP</i> address and port were supplied beforehand, namely the <i>IP</i> address `192.168.0.100` with port `80`. When using this filter, it becomes obvious that the trace contains alot of of records which do not belong to the attack. After applying the filter only <b>67071</b> records remain.

Using the statistics of Wireshark, the <i>IPv4</i> addresses which were active in the attack can bee seen. To do this, the Wireshark statistics were used: `Statistics > IPv4 Statistics > Source and Destination Addresses`. Here the filter described above can be appllied. The result is a list of <b>17</b> addresses which took part in the attack. the same can be done for <i>IPv6</i> addresses by going to `Statistics > IPv6 Statistics > Source and Destination Addresses`. Here it can be seen that the list is empty, meaning that no <i>IPv6</i> addresses were used in the attack.

When looking at the <i>IPv4</i> statistic, the list can be sorted on the count column. Doing this, the <b>3</b> <i>IP</i> addresses which are responsible for the most packets can be determined. As can be seen in the chart below, these <b>3</b> <i>IP</i> addresses together were responsible for just over <b>50%</b> of the attack traffic.

```python
top_3_source_ip = pd.read_csv('data/top_3_source_ip.csv')
pie_chart(top_3_source_ip['count'], top_3_source_ip['IP address'], 'Top 3 IP addresses')
```

On this site a <i>MAC</i> address can be entered and the coresponding vendor will be shown. Using the <i>MAC</i> addresses of the attackers, the following distribution was found.

Although the destination port for all requests is the same, namely <b>80</b>, the source ports are not. The source ports used lie in the range from <b>13051</b> up to <b>64827</b>. Not all of the ports in this range are used, there are a lot of small gaps in there.

Not only the <i>IP</i> addresses and vendors were looked at, but also the data rate, the frequency of <i>TCP</i> flags, and the frequency of packet length was looked at.   

First, the data rate. Using Wireshark an graph of the data rate was generated (`Statistics > I/O Graphs`). In this we show a single line which contains the byte rate on the y axis. The display filter is the same filter as the one used to only show traffic for the attack: `tcp.dstport == 80 && ip.dst == 192.168.0.100`. An SMA period of `10 interval SMA` was used to smoothen the graph.
![Wireshark Data rate](screenshots/data_rate.png)

As can be seen there is a peak at 36 seconds. The data rate here is <i>718300 bytes/sec</i>. By making a simple calculation it can be converted to <i>Mbits/sec</i>:  

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;718300 <i>bytes/sec</i> * 8 = 5746400 <i>bits/sec</i>  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;5746400 <i>bits/sec</i> / 10^6 = 5.7464 <i>Mbits/sec</i>
